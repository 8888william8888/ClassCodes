## 模型的表示
    
- 模型：
    - 经验建模：
        - $y = w x  +  b$
        - $y = w_1x_1 + w_2x_2 + \dots + w_{n}x_{n} + b$
        
        - $y = W X + b$ : 向量乘积：内积、点积
            - $W = (w_1,w_2,\dots,w_n)$
            - $X = (x_1,x_2,\dots,x_n)$
            
        - 另外表示方式：
            - $X_i=(x_{1i},x_{2i},\dots, x_{ni}, 1)$
            - $W=(w_1,w_2,\dots, w_n, b)$
            - $y_i= WX_i$    : $Xi$就是第$i$个样本, 就是一条直线
            
- 定义：
    - 约定俗成


- 线性：
    - 直线，平面，

````python
# y = xw  + b

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg   as la   

# w = np.array([
#     [0.3]
# ])
w = 0.3
b = 4

line = lambda  x:  w * x + b

figure  = plt.figure('直线', figsize=(8, 6))
ax = figure.add_axes([0.1, 0.1, 0.8, 0.8])
ax.set_xlim(0,5)
ax.set_ylim(0,5)

x = np.linspace(0, 5 ,100)
# x = np.reshape(x, (100,1))
y = line(x)

print(x.shape)
print(y.shape)

ax.scatter(x, y, s=9, c='r')

figure.show(warn=False)
````

## 模型的推导

- 线性永远不可能是理想的线性模型，肯定存在误差：
    - 年龄不存在误差，身高存在测量误差，误差表示：$\epsilon$
    
    
    
    
- 误差情况下线性模型：线性模型的误差模型。
    - $y_i = x_i W  + \epsilon_{i}$
    
         - 其中$\epsilon_i$是第$i$个样本的误差

- 误差的原则：
    - 误差越小，概率越大（产生小误差概率越大是最好的情况）

- 误差的概率模型：
    - 假设：误差遵循高斯分布。（最符合人类的直观期望，其实还有其他概率规则）

- 概率的计算公式
    - 高斯分布：中心点：均值点$\mu$，离散度：偏差、方差$\sigma$
    - 高斯分布表示：$N(\mu,\sigma)$
    
    - $N(x, \mu,\sigma) = \dfrac{1}{\sqrt{2\pi} \sigma}  exp( - \dfrac{(x- \mu) ^2 }{2 \sigma ^ 2}  ) $

- 假设高斯分布：
    - $\mu = 0$
    - $\sigma = 1$
    
    - 标准正态分布：标准高斯分布
    
    -$N(x) = \dfrac{1}{\sqrt{2\pi}}  exp( - \dfrac{x ^2 }{2}  ) $

- 使用高斯公司表示误差

    - $N(\epsilon_i) = \dfrac{1}{\sqrt{2\pi}}  exp( - \dfrac{\epsilon_i ^2 }{2}  ) $

## 补充-误差的概率计算

- 使用matplotlib演示误差的概率。

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

figure = plt.figure('标准正态概率函数',figsize=(8, 4))
ax = figure.add_axes([0.1, 0.1, 0.8,0.8])
ax.set_ylim(0, 0.5)
n = lambda x :  (1.0/ np.sqrt(np.pi * 2.0)) * np.exp( - x**2 / 2.0 )
x = np.linspace(-5,5, 100)
y = n(x)

ax.plot(x, y, color=(0, 0, 1, 1))

# 计算：产生0误差的概率，产生0.001的误差的概率
y_0 = n(10)
y_0001= n(0.0001)
print(y_0, y_0001)
ax.scatter(10, y_0, c='r', marker='o', s = 64)
ax.scatter(0.0001, y_0001, c='r', marker='o', s= 64)

figure.show(warn=False)
````

## 线性回归的优化模型

- 找到一条直线，是的已知数据的误差最小，概率最大。

## 把误差模型应用到线性模型

- 表达式代入

- $N(\epsilon_i) = p(\epsilon_i | x_i, W) = \dfrac{1}{\sqrt{ 2 \pi} \sigma} exp(- \dfrac{(y_i - x_i W)^2}{2})$

    - 其中$i$表示第$i$个样本的误差概率

- 误差越小，概率越大。

- 50个样本的误差同时发生的概率怎么计算。
    - 同时发生的概率表达。所有的概率相乘。
    
- 所有样本的误差发生概率
    
    - $L(W) =  \prod \limits_{i=1} ^{n} N(\epsilon_i) = \prod \limits_{i=1}^{n} p(\epsilon_i | x_i, W)$
    
    - 其中的n表示所有的样本


- 上面的函数
     - 称为：极大(最大）似然概率函数
     
     
- 问题的变化：
    - 通过算法找到一个$W$使得$L(W)$的值最大（概率最大，误差最小）。
     

1.  **选在线性模型是根据经验。**
2. **线性模型的数学表达：** $ y = Wx $ = $ w_1x_1 + w2x_2 + \dots +w_nx_n  + b$  
3. **所有的已知点不完全在直线上，存在误差，误差的表达：$y = W x + \epsilon$**
4. **误差的规律满足加入：$\epsilon$满足高斯分布（正态分布$\to$标准正态分布）**
5. **根据高斯概率函数，已知误差，就可以计算出误差出现的概率。**
6. **如果训练样本已知，就计算出所有训练样本同时出现误差的概率：概率似然函数**
7. **是否存在W使得濡染概率函数的值最大。**


## 求解最大值问题


$L(W) =  \prod \limits_{i=1} ^{n} N(\epsilon_i) = \prod \limits_{i=1}^{n} p(\epsilon_i | x_i, W)$

 $N(\epsilon_i) = p(\epsilon_i | x_i, W) = \dfrac{1}{\sqrt{ 2 \pi} \sigma} exp(- \dfrac{(y_i - x_i W)^2}{2})$


$L(W) =\prod \limits_{i=1}^{n}  \dfrac{1}{\sqrt{ 2 \pi} \sigma} exp(- \dfrac{(y_i - x_i W)^2}{2})$

- 等式两边求对数

    - $ln L(W) =ln  (\prod \limits_{i=1}^{n}  \dfrac{1}{\sqrt{ 2 \pi} \sigma} exp(- \dfrac{(y_i - x_i W)^2}{2}) )$

- $l(W) = \sum \limits_{i=1} ^{n}  ln  (  \dfrac{1}{\sqrt{ 2 \pi} \sigma} exp(- \dfrac{(y_i - x_i W)^2}{2} )$

- $l(W) = \sum \limits _ {i=1} ^ {n}  ln  \dfrac{1}{\sqrt{2 \pi} \sigma}    +  ln exp ( - \dfrac{(y_i- x_iW)^2}{2}) $

- $l(W) = n  ln \dfrac{1}{\sqrt{2 \pi}}  - \dfrac{1}{2} \sum \limits _{i=1}^n ln (exp (y_i - x_i W) ^2)$

- $l(W) = n  ln \dfrac{1}{\sqrt{2 \pi}}  - \dfrac{1}{2} \sum \limits _{i=1}^n (y_i - x_i W) ^2$   最大

 - $\dfrac{1}{2} \sum \limits _{i=1}^n (y_i - x_i W) ^2 $ 找到W使得公式值最小

## 求方差的最小值

-  $J(W) = \dfrac{1}{2} \sum \limits _{i=1}^n (y_i - x_i W) ^2 $ 找到W使得公式值最小

- 损失函数（残差函数）

- 上面的样本损失函数按照50个样本，写成矩阵表示的表示模式（求和表示内积）

$J(W) = \dfrac{1}{2} (Y-XW)^T (Y-XW)$

- 最小值的推导
    - 求导数，取导数的0值点。 0值点就是最小值点。
    - 最小二乘法

- $\nabla _W J(W)  = \nabla_W \dfrac{1}{2} (Y-XW)^T (Y-XW) $

- $\nabla _ W J(W) =  \nabla _W \dfrac{1}{2} (  Y^T Y - Y^T XW  - W^T X^TY + W^TX^T XW ) $

 - $\nabla _ W J(W) =  \dfrac{1}{2}  (  2 X^T XW - X^TY - (Y^TX)^T )$

-  $\nabla _ W J(W) = X^TXW -X^TY $

- 求最小：
    - $X^TXW -X^TY = 0 $

- $W = ( X^TX)^{-1} X^TY$

## numpy实现


-  个体数据 $\to$ 向量，矩阵

- $W = ( X^TX)^{-1} X^TY$

````python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np             
import numpy.linalg as la     # 线性代数

# 1. 加载
X_DATA = np.loadtxt('ex2x.dat')   # 年龄
Y_DATA = np.loadtxt('ex2y.dat')   # 身高

# 对X数据处理（每个数据添加1，对应权重），Y数据处理
# 处理X数据
X = np.zeros( shape= (X_DATA.shape[0] ,2 ) , dtype=np.float)
X[:, 0] = X_DATA
X[:, 1] = 1

# 处理Y处理
Y = Y_DATA.reshape(Y_DATA.shape[0], 1)

# 2. 算法，求W 
S1 = np.matmul(X.T, X)
S2 = la.inv(S1)     # 存在问题：（矩阵没有逆矩阵）
S3 = np.matmul(S2, X.T)
S4 = np.matmul(S3 ,Y)
print(S4)
# 3. 可视化效果（样本点，回归直线）
# 绘制样本点
figure  = plt.figure('年龄与身高', figsize=(12, 6))
ax = figure.add_axes( [0.1, 0.1, 0.8, 0.8] )
ax.set_xlim(0, 9)
ax.set_ylim(0.3, 1.5)
ax.scatter(X_DATA,Y_DATA, s=9, c='b', marker='.')

# 绘制直线
X_POS = np.linspace(1.5, 8.5, 100, dtype=np.float)
X_PRA = np.zeros( (X_POS.shape[0],2) ,dtype=np.float)
X_PRA[:,0] = X_POS
X_PRA[:,1] =1 

# 计算直线的Y坐标
W = S4
Y_PRA = np.matmul(X_PRA, W)

# 推荐：Y_PRA做一个形状变换
ax.plot(X_POS, Y_PRA, color=(1, 0, 0, 1), label='线性回归')

ax.legend()

figure.show(warn=False)
````

## sklearn的验证实现

````python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np             
import numpy.linalg as la     # 线性代数
import sklearn.linear_model as lm

# 1. 加载
X_DATA = np.loadtxt('ex2x.dat')   # 年龄
Y_DATA = np.loadtxt('ex2y.dat')   # 身高

# 对X数据处理（每个数据添加1，对应权重），Y数据处理
# 处理X数据
X = X_DATA.reshape(-1, 1)
Y = Y_DATA
# 2. 算法，求W 
reg = lm.LinearRegression()
reg.fit(X, Y)
P1 = reg.coef_
P2 = reg.intercept_ 
print(P1, P2)
````
