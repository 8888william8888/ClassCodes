# 时序图说明

- UML图一共9种，分成5静4动：
    - 静态图：
        - 用例图
        - 对象图
        - 类图
        - 部署图
        - 组件图
        
    - 动态图：
        - 时序图
        - 活动图
        - 状态图
        - 协作图

- 时序图(Sequence Diagram)，是一种UML中的一种动态图，用来说明对象的时序调用交互。
    - 它通过描述对象之间发送消息的先后顺序显示多个对象之间的调用关系与顺序。
    - 时序图用来确认与现实对象协作完成的业务流程。
    

# 时序图的基本元素

- 时序图时会涉及7种元素
    - 角色(Actor)，也称参与者。
    - 对象(Object)，类的具体实例。
    - 时间线(LifeLine)
    - 控制焦点(Activation)
    - 消息(Message)
    - 自关联消息
    - 组合片段，用来表示复杂的条件，并行执行等表达（不介绍）。

## 消息

- 表示对象之间发送的消息或者一次调用；使用箭头线表示。

- 消息分成两种：
    - 调用消息与调用返回
        - 调用使用：实线 + 实心箭头
        - 返回使用：虚线 + < 箭头
        
- 对象自己发送给自己的消息称为自关联消息
