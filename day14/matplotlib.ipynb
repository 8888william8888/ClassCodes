# 2. Figure的容器结构

- 对象的构建的数据设置：
    - |-构造器/函数参数设置
    - |-set/get设置/获取参数
    - |- rcParams\['xxx'\]=值 

```python
% matplotlib inline

# 1.构建对象Figure（不建议自己使用构造器构建）
figure = plt.figure(
    num='图形',
    figsize=(5.5, 4),
    dpi=100,
    facecolor=(1,1,1,1),
    edgecolor=(1,1,1,1),
    frameon=True,
    linewidth=0
)

# 2. 构建坐标轴
ax = plt.Axes(
    fig=figure,
    rect=[0.1, 0.1, 0.8, 0.8],
    frameon=True,
    fc=(1,1,1,1),
    label='标签',
    title='标题'
)

#ax.set_xlim(left=2,right=-2)
#ax.set_xbound(lower=-5,upper=5)
#控制刻度：自动（算法）、手动，算法
#ax.set_xmargin(0.5)
#ax.set_xscale('log')
'''
ax.set_xticks(
    ticks=[0,2,3.5,4,5,7],
    minor=False
)
ax.set_xticklabels(
    labels=['X','Y','Z','A'],
    fontdict={
        'color': (1,0,0,1),
        'rotation': 45
    }
)
ax.set_xlabel(
    'x轴',
    labelpad=-40
)
'''
import numpy as np 
# 3. 绘制图形
line=plt.Line2D(
    xdata=np.linspace(-10.0,10.0,21),
    ydata=np.random.uniform(-1,1,size=21),
    linewidth=1,
    linestyle='-',   #(0.1, (10, 5, 3, 10) ),
    color=(1,0,0,1),
    solid_capstyle='butt',
    solid_joinstyle='round',
    marker='.',
    markersize=20,
    markeredgewidth=2,
    markerfacecolor=(1,1,0,1),
    markeredgecolor=(0,0,1,1),
    markerfacecoloralt=(0,1,1,1),
    fillstyle='top',
    markevery=1
    
)

ax.grid(b=True)
ax.set_xlim(left=-10,right=10)
ax.set_ylim(bottom=-1, top=1)

xaxis=ax.get_xaxis()
from matplotlib.ticker import *

ml=MultipleLocator(5)
xaxis.set_major_locator(ml)

ml2=MultipleLocator(0.2)
xaxis.set_minor_locator(ml2)

pf=PercentFormatter(1, decimals=2, symbol='‰$a2+A_{11}$', is_latex=True)
xaxis.set_major_formatter(pf)

figure.add_axes(ax)

ax.add_artist(line)
#4.显示
figure.show(warn=False)
```
